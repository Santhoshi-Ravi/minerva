# **get summary or answer**

```python
# Use regular expression to find the numeric value
import re
# Define a regular expression pattern to match the last set of numbers in the text
def get_answer(resp):
  pattern = r'([-+]?\d*\.\d+|[-+]?\d+)(?!\d.*\d)'

  # Search for the pattern in the output
  matches = re.findall(pattern, str(resp))

  if matches:
      # Extract the last numbers from the list of matches
      last_numbers = matches[-1]
  return last_numbers
```

# LLAMA

```python
from transformers import AutoTokenizer
import transformers
import torch

model = "meta-llama/Llama-2-7b-chat-hf" # meta-llama/Llama-2-7b-hf

tokenizer = AutoTokenizer.from_pretrained(model, use_auth_token=True)
```

```python
from transformers import pipeline

llama_pipeline = pipeline(
    "text-generation",  # LLM task
    model=model,
    torch_dtype=torch.float16,
    device_map="auto",
)
```

```python
def get_llama_response(prompt: str) -> None:
    """
    Generate a response from the Llama model.

    Parameters:
        prompt (str): The user's input/question for the model.

    Returns:
        None: Prints the model's response.
    """
    sequences = llama_pipeline(
        prompt,
        do_sample=True,
        top_k=10,
        num_return_sequences=1,
        eos_token_id=tokenizer.eos_token_id,
        max_length=100000,
    )
    out="Chatbot:", sequences[0]['generated_text']
    return out



# prompt = math_ques[0]
prompt="What is Total of 0.06 and -1977321735."

print(get_llama_response(prompt))
```

ACCURACY

```python
accuracy=0
for math_q,math_a in zip(math_ques,math_ans):
    prompt= math_q
    resp=get_llama_response(prompt+" Lets Think step by step,Please.")
    # print(resp)
    # if float(get_answer(math['answer'])) != float(get_answer(resp)):
    #   resp=get_llama_response(str(resp)+" Are you sure?Thats not right. Lets Rethink step by step")
    # print(resp)
    # break
    #COT+self-consistency
    resp_self_const=get_llama_response(prompt+"This is your answer:"+resp[1]+"Are you sure?Give me a final answer")
    if float(get_answer(math_a)) == float(get_answer(resp_self_const)):
      accuracy=accuracy+1

print(accuracy)
```
